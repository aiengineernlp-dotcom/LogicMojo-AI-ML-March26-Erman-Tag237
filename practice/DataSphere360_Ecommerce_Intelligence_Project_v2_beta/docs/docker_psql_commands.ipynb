```python
1- Trouver le nom de votre conteneur
docker ps

2. Entrer dans le conteneur et ouvrir PostgreSQL
docker exec -it NOM_DU_CONTENEUR psql -U NOM_UTILISATEUR -d NOM_BASE_DE_DONNEES
exemple : docker exec -it lead-capture-db-1 psql -U postgres -d postgres

3. Les commandes pour lister les tables
Une fois que votre invite de commande affiche postgres=#, utilisez ces raccourcis clavier PostgreSQL :
\dt : Liste toutes les tables disponibles (c'est ici que vous verrez si la table lead existe).
\d lead : Affiche la structure, les colonnes et les types de la table lead.
\q : Permet de quitter l'interface PostgreSQL et revenir à votre terminal.
a
```
